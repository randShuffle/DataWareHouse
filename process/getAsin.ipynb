```python
import numpy as np
import pandas as pd

columns=['productId','userId','profileName','helpfulness','score','time','summary','text']

data=[]
single_data=[]
i=0
col_len=len(columns)
with open("data/movies.txt",'r',errors="ignore") as f:
    while True:
        line=f.readline()
        #at the end of the file
        if not line:
            break
        #get the line without space or '\n' etc
        line=line.strip()
        #skip the empty line
        if len(line)==0:
            continue
        else:   
            prefix_col=line.split(':',1)[0].strip().split('/',1)[-1]
            if prefix_col not in columns:
                if len(single_data)==0:
                    print("error")
                else:
                    single_data[-1]+=(" "+line)
            else:    
                single_data.append(line.split(':',1)[-1].strip())
                if len(single_data)==col_len:
                    data.append(single_data.copy())
                    single_data.clear()
                    i+=1
                    #just for observe the process
                    if i%10000==0:
                        print(i)
                   
            

        

      


df=pd.DataFrame(data=data,columns=columns)


df.to_csv("data/movies.csv",index=False)

print("csv end")


```

```python
import pandas as pd
from itertools import combinations, product
from collections import defaultdict

# 创建一个示例的 DataFrame（假设名为 df），包含 'actor' 和 'director' 列，其中 'director' 列是一个列表
data = {
  'actor': [
    ['Actor A', 'Actor B', 'Actor C'],
    ['Actor A', 'Actor B'],
    ['Actor A', 'Actor D', 'Actor E'],
    ['Actor B', 'Actor D'],
    ['Actor C', 'Actor E'],
    ['Actor D', 'Actor E']
  ],
  'director': [
    ['Director X', 'Director Y'],
    ['Director Y'],
    ['Director X', 'Director Z'],
    ['Director Z'],
    ['Director Y'],
    ['Director Z']
  ]
}

df = pd.DataFrame(data)

# 创建一个 defaultdict 用于存储合作次数
collaborations = defaultdict(int)

# 遍历 DataFrame 中的每一行，统计演员、导演以及演员与导演之间的合作次数
for i, row in df.iterrows():
  actors_list = row['actor']
  directors_list = row['director']

  # 统计演员之间的合作次数
  for pair in combinations(actors_list, 2):
    collaborations[tuple(sorted(pair))] += 1

  # 统计导演之间的合作次数
  for director_pair in combinations(directors_list, 2):
    collaborations[tuple(sorted(director_pair))] += 1

  # 统计演员与导演之间的合作次数
  for actor, director in product(actors_list, directors_list):
    collaborations[(actor, director)] += 1

# 输出每对演员、导演以及演员与导演之间的合作次数
for pair, count in collaborations.items():
  print(f"合作次数 {pair}: {count}")
```

```python
from datetime import datetime
import pandas as pd
from itertools import combinations,product
from collections import defaultdict
import ast
myMovieList = []
myActorDict = {}
myActorList = []
myDirectorDict = {}
myDirectorList = []
myGenresDict = {}
myGenresList = []
myFormatDict={}
myFormatList=[]
myTimeDict={}
myTimeList=[]

myMovieReleaseTimeRelationList = []
myGenresRelationList = []
myFormatRelationList=[]
myDirectorRelationList = []
myStarringRelationList = []
mySupportingRelationList = []
myAACooperationRelationList = []
myADCooperationRelationList = []
myDDCooperationRelationList = []

outputFolder = "../load/"
myMoviePath = outputFolder + "movies.csv"
myActorPath = outputFolder + "actors.csv"
myDirectorPath = outputFolder + "directors.csv"
myFormatPath = outputFolder + "formats.csv"
myTimePath = outputFolder + "times.csv"
myGenresPath = outputFolder + "genres.csv"

myMovieReleaseTimeRelationPath = outputFolder + "movie_release_time_relation.csv"
myGenresRelationPath = outputFolder + "genres_relation.csv"
myFormatRelationPath = outputFolder + "formats_relation.csv"
myDirectorRelationPath = outputFolder + "director_relation.csv"
myStarringRelationPath = outputFolder + "starring_relation.csv"
mySupportingRelationPath = outputFolder + "supporting_relation.csv"
myAACooperationRelationPath = outputFolder + "aa_cooperation_relation.csv"
myADCooperationRelationPath = outputFolder + "ad_cooperation_relation.csv"
myDDCooperationRelationPath = outputFolder + "dd_cooperation_relation.csv"

converters={'genre': ast.literal_eval,
            'actor': ast.literal_eval,
            'director': ast.literal_eval,
            'format': ast.literal_eval,
            'language': ast.literal_eval,}
df_movie_merge=pd.read_csv('../data/movies-after-merge2.csv',index_col=0,converters=converters).head(10000)


for index, row in df_movie_merge.iterrows():
   
    #movie节点
    movie_id = len(myMovieList)
    title = row['title']
    runTime = row['run_time']
    #score = row['score']
    #isPositive=row['isPositive']
    #commentNumber = row['commentNumber']
    asin=row['asin']
    #myMovieList.append([movie_id, title, runTime, score, isPositive, commentNumber,asin])
    myMovieList.append([movie_id, title, runTime,asin])
    
    
    #time节点
    time=row['release_date']
    time = time.strip()
    tmp = len(myTimeDict)
    #time not exist
    if time not in myTimeDict.keys():
        myTimeDict[time] = tmp
        date_object = datetime.strptime(time, '%Y-%m-%d')
        myTimeList.append([tmp,time,date_object.year,date_object.month,date_object.day,date_object.weekday(),(date_object.month - 1) // 3 + 1])
    else:
        tmp = myTimeDict[time]

    #movie-time relation
    myMovieReleaseTimeRelationList.append([movie_id, tmp])

    #genre节点
    for genres in row['genre']:
        genres = genres.strip()
        tmp = len(myGenresDict)
        #genre not exist
        if genres not in myGenresDict.keys():
            myGenresDict[genres] = tmp
            myGenresList.append([tmp,genres])
        else:
            tmp = myGenresDict[genres]

        #movie-genre relation
        myGenresRelationList.append([movie_id, tmp])

    #format节点
    for format in row['format']:
        format = format.strip()
        tmp = len(myFormatDict)
        #format not exist
        if format not in myFormatDict.keys():
            myFormatDict[format] = tmp
            myFormatList.append([tmp,format])
        else:
            tmp = myFormatDict[format]

        #movie-format relation
        myFormatRelationList.append([movie_id, tmp])


    #actor节点
    for i,actor in enumerate(row['actor']):
        actor = actor.strip()
        tmp = len(myActorDict)
        #actor not exist
        if actor not in myActorDict.keys():
            myActorDict[actor] = tmp
            myActorList.append([tmp,actor])
        else:
            tmp = myActorDict[actor]

        #star relation
        myStarringRelationList.append([movie_id, tmp]) if i<2 else  mySupportingRelationList.append([movie_id, tmp])
        
    
    #director节点
    for director in row['director']:
        director = director.strip()
        tmp = len(myDirectorDict)
        #director not exist
        if director not in myDirectorDict.keys():
            myDirectorDict[director] = tmp
            myDirectorList.append([tmp,director])
        else:
            tmp = myDirectorDict[director]

        #movie-director relation
        myDirectorRelationList.append([movie_id, tmp])


#cooperation relation


# 创建一个 defaultdict 用于存储合作次数
collaborations = defaultdict(int)
for actors_list in df_movie_merge['actor']:
  actors_list=[myActorDict[name] for name in actors_list]
  for pair in combinations(actors_list, 2):
    # 对每一对演员进行计数
    collaborations[tuple(sorted(pair))] += 1

for pair, count in collaborations.items():
    myAACooperationRelationList.append([pair[0],pair[1],count])



collaborations = defaultdict(int)
for director_list in df_movie_merge['director']:
  director_list=[myDirectorDict[name] for name in director_list]
  for pair in combinations(director_list, 2):
    # 对每一对演员进行计数
    collaborations[tuple(sorted(pair))] += 1

for pair, count in collaborations.items():
    myDDCooperationRelationList.append([pair[0],pair[1],count])



collaborations = defaultdict(int)
# 遍历 DataFrame 中的每一行，统计演员、导演以及演员与导演之间的合作次数
for actors_list, directors_list in zip(df_movie_merge['actor'], df_movie_merge['director']):
    actors_list=[myActorDict[name] for name in actors_list]
    directors_list=[myDirectorDict[name] for name in directors_list]
  # 统计演员与导演之间的合作次数
    for actor, director in product(actors_list, directors_list):
        collaborations[(actor, director)] += 1

for pair, count in collaborations.items():
    myADCooperationRelationList.append([pair[0],pair[1],count])

#保存到csv
#df_movie=pd.DataFrame(myMovieList,columns=['movie_id','title','runTime','score','isPositive','commentNumber','asin'],data=myMovieList)
df_movie=pd.DataFrame(columns=['movie_id','title','runTime','asin'],data=myMovieList)
df_movie.to_csv(myMoviePath,header=True)

df_actor=pd.DataFrame(columns=['actor_id','actor'],data=myActorList)
df_actor.to_csv(myActorPath,header=True)

df_director=pd.DataFrame(columns=['director_id','director'],data=myDirectorList)
df_director.to_csv(myDirectorPath,header=True)

df_format=pd.DataFrame(columns=['format_id','format'],data=myFormatList)
df_format.to_csv(myFormatPath,header=True)

df_time=pd.DataFrame(columns=['time_id','time','year','month','day','weekday','season'],data=myTimeList)
df_time.to_csv(myTimePath,header=True)

df_movie_release_time_relation=pd.DataFrame(columns=['movie_id','time_id'],data=myMovieReleaseTimeRelationList)
df_movie_release_time_relation.to_csv(myMovieReleaseTimeRelationPath,header=True)

df_genres=pd.DataFrame(columns=['genres_id','genres'],data=myGenresList)
df_genres.to_csv(myGenresPath,header=True)

df_genres_relation=pd.DataFrame(columns=['movie_id','genres_id'],data=myGenresRelationList)
df_genres_relation.to_csv(myGenresRelationPath,header=True)

df_format_relation=pd.DataFrame(columns=['movie_id','format_id'],data=myFormatRelationList)
df_format_relation.to_csv(myFormatRelationPath,header=True)

df_director_relation=pd.DataFrame(columns=['movie_id','director_id'],data=myDirectorRelationList)
df_director_relation.to_csv(myDirectorRelationPath,header=True)

df_starring_relation=pd.DataFrame(columns=['movie_id','actor_id'],data=myStarringRelationList)
df_starring_relation.to_csv(myStarringRelationPath,header=True)

df_supporting_relation=pd.DataFrame(columns=['movie_id','actor_id'],data=mySupportingRelationList)
df_supporting_relation.to_csv(mySupportingRelationPath,header=True)

df_aa_cooperation_relation=pd.DataFrame(columns=['actor_id','actor_id','count'],data=myAACooperationRelationList)
df_aa_cooperation_relation.to_csv(myAACooperationRelationPath,header=True)

df_dd_cooperation_relation=pd.DataFrame(columns=['director_id','director_id','count'],data=myDDCooperationRelationList)
df_dd_cooperation_relation.to_csv(myDDCooperationRelationPath,header=True)

df_ad_cooperation_relation=pd.DataFrame(columns=['actor_id','director_id','count'],data=myADCooperationRelationList)
df_ad_cooperation_relation.to_csv(myADCooperationRelationPath,header=True)
```

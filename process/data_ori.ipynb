```python
#Harry Potter溯源分析
import pandas as pd
df=pd.read_csv('../data/movies-relationship.csv',index_col=0)
result_df = df[df.index.str.contains('harry potter', case=False)]
result_df
```

```python
result_df.to_csv('../data/harry_potter.csv')
```

```python
import ast
converters={'asin':ast.literal_eval,
            'genre': ast.literal_eval,
            'actor': ast.literal_eval,
            'director': ast.literal_eval,
            'format': ast.literal_eval,
            'language': ast.literal_eval,}
df=pd.read_csv('../data/harry_potter.csv',index_col=0,converters=converters)
df.tail(3)

```

```python
#统计有多少版本
df['format_num']=df['format'].apply(lambda x:len(x))
#统计有多少网页
df['total_num']=df['asin'].apply(lambda x:len(x))

final_ans=df[['format_num','total_num']]
final_ans.to_csv('../data/harry_potter_final_ans.csv')
```
